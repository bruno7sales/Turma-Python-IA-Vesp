# PANDAS E SUAS ULTILIDADES:

 

<h4>Pandas  uma biblioteca de software criada para a linguagem Python para manipulao e anlise de dados. Em particular, oferece estruturas e operaes para manipular tabelas numricas e sries temporais.  software livre sob a licensa licena BSD. O nome  derivado do termo ingls "panel data"(dados em painel), um termo usado em estatstica e econometria para conjunto de dados que incluem vrias unidades amostrais (indivduos, empresas, etc) acompanhadas ao longo do tempo.</h4>
